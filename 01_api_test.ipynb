```python
import os
from dotenv import load_dotenv
from riotwatcher import RiotWatcher, LolWatcher, ApiError

load_dotenv(dotenv_path="../.env", override=True)
api_key = os.getenv("RIOT_API_KEY")

if not api_key:
    raise ValueError("RIOT_API_KEY not found in .env file")

print("Key prefix:", api_key[:10])
print("Key length:", len(api_key))

riot_watcher = RiotWatcher(api_key)
lol_watcher = LolWatcher(api_key)

print("API client initialized successfully")
```

```python
riot_id_name = "Hide on bush"
riot_id_tag = "KR1"
account_region = "asia"

try:
    account_data = riot_watcher.account.by_riot_id(account_region, riot_id_name, riot_id_tag)
    puuid = account_data["puuid"]
    print("PUUID fetched successfully")
    print(puuid)
except ApiError as err:
    print(f"API error: {err.response.status_code}")
```

```python
platform_routing = "kr"

try:
    summoner_data = lol_watcher.summoner.by_puuid(platform_routing, puuid)
    print("Summoner profile fetched successfully")
    print("Summoner Level:", summoner_data["summonerLevel"])
    print("Profile Icon ID:", summoner_data["profileIconId"])
except ApiError as err:
    print(f"API error: {err.response.status_code}")
```
