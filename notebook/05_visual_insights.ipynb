```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/faker_recent_matches.csv")
champion_summary = pd.read_csv("../data/processed/champion_summary.csv")

print("match data shape:", df.shape)
print("champion summary shape:", champion_summary.shape)
```

```python
df["win"] = df["win"].astype(int)
df["game_duration_min"] = df["game_duration"] / 60
df["game_creation"] = pd.to_datetime(df["game_creation"], unit="ms", errors="coerce")
df["kda"] = df["kda"].replace([np.inf, -np.inf], np.nan)

print(df[["champion", "win", "kda", "gold_earned", "total_damage_to_champions"]].head())
```

```python
top_champions = df["champion"].value_counts().head(10)

plt.figure(figsize=(10, 5))
top_champions.plot(kind="bar", color="steelblue")
plt.title("Top 10 Most Played Champions")
plt.xlabel("Champion")
plt.ylabel("Match Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
champion_summary_filtered = champion_summary[champion_summary["matches"] >= 3].copy()
champion_summary_filtered = champion_summary_filtered.sort_values("win_rate", ascending=False)

plt.figure(figsize=(10, 5))
plt.bar(champion_summary_filtered["champion"], champion_summary_filtered["win_rate"], color="seagreen")
plt.title("Win Rate by Champion (Min. 3 Matches)")
plt.xlabel("Champion")
plt.ylabel("Win Rate")
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
win_counts = df["win"].value_counts().sort_index()
win_counts.index = ["Loss", "Win"]

plt.figure(figsize=(6, 4))
win_counts.plot(kind="bar", color=["tomato", "seagreen"])
plt.title("Win vs Loss Count")
plt.xlabel("Result")
plt.ylabel("Match Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["kda"].dropna(), bins=20, edgecolor="black", color="mediumpurple")
plt.title("KDA Distribution")
plt.xlabel("KDA")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["gold_earned"], df["total_damage_to_champions"], alpha=0.7, color="darkorange")
plt.title("Gold Earned vs Damage to Champions")
plt.xlabel("Gold Earned")
plt.ylabel("Damage to Champions")
plt.tight_layout()
plt.show()
```

```python
colors = df["win"].map({0: "tomato", 1: "seagreen"})

plt.figure(figsize=(8, 6))
plt.scatter(df["kills"], df["deaths"], c=colors, alpha=0.7)
plt.title("Kills vs Deaths by Match Result")
plt.xlabel("Kills")
plt.ylabel("Deaths")
plt.tight_layout()
plt.show()
```

```python
position_counts = df["individual_position"].fillna("Unknown")
position_counts = position_counts[position_counts != "Invalid"]
position_counts = position_counts.value_counts()

plt.figure(figsize=(8, 5))
position_counts.plot(kind="bar", color="slateblue")
plt.title("Position Distribution (Valid Roles Only)")
plt.xlabel("Position")
plt.ylabel("Match Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
daily_matches = df.set_index("game_creation").resample("D").size()

plt.figure(figsize=(10, 5))
daily_matches.plot(color="teal")
plt.title("Matches Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Matches")
plt.tight_layout()
plt.show()
```

## Visual Insights

- The recent match history is concentrated among a relatively small set of frequently played champions.
- Match outcomes show a visible difference between wins and losses across the sampled games.
- KDA varies substantially across matches, indicating uneven game-to-game performance and several strong outlier performances.
- Gold earned appears to have a positive relationship with damage dealt to champions.
- Position labels include some noisy values such as `Invalid`, which reflects raw match data quality rather than a pipeline error.
- The exported visual summary table can support future dashboards, deeper analysis, or portfolio reporting.
